# Scattering of a plane wave on a rectangular potential barrier

Necessary imports and settings

```python
import warnings
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (20,12)
plt.matplotlib.rcParams.update({'font.size': 22})
warnings.simplefilter("ignore", category=RuntimeWarning)
```

Function transmission_coefficient returns coefficient of transmission through the rectangular potential barrier. This is the only kind of barrier that can be calculated analytically. 

```python
def transmission_coefficient(a, E):
    
    if E == 1.0:
        E += 10e-7

    if E > 1.0:
        return 1.0 / (1.0 + ((np.sin(np.sqrt(2.0*(E-1.0)) * a))**2) / (4.0*E*(E-1.0)))
    else:
        return 1.0 / (1.0 + ((np.sinh(np.sqrt(2.0*(1.0-E)) * a))**2) / (4.0*E*(1.0-E)))
```

When energy equals one, a small number is added to avoid dividing by zero.

## Function of energy

```python
#these 3 values can be safely changed
potential_width = 8
energy_range = 4
step = 0.01

# x axis
Energy = np.arange(0,energy_range+step,step)

# y axis
TransmissionCoefficients = []

for energy in Energy:
    TransmissionCoefficients.append(transmission_coefficient(potential_width, energy))
    
#plot settings
plt.ylim([-0.01,1.01])
plt.xlim([0,energy_range])
plt.ylabel('Transmission')
plt.xlabel('Energy')
plt.grid(True)
plt.plot(Energy, TransmissionCoefficients)
plt.show()
```

## Function of potential barrier width

```python
#these 3 values can be safely changed
energy = 1.1
potential_width_range = 30
step = 0.01

# x axis
Potential_width = np.arange(0,potential_width_range+step,step)

#y axis
TransmissionCoefficients = []

for potential_width in Potential_width:
    TransmissionCoefficients.append(transmission_coefficient(potential_width, energy))
    
#plot settings
plt.ylim([-0.01,1.01])
plt.xlim([0,potential_width_range])
plt.ylabel('Transmission')
plt.xlabel('Energy')
plt.grid(True)
plt.plot(Potential_width, TransmissionCoefficients)
plt.show()
```
